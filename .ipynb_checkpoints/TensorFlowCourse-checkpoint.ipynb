```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier  # Importation du modèle Random Forest
from PIL import Image
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Load Data Set
(train_X, train_y), (test_X, test_y) = mnist.load_data()

# Train data set IMGS
image_1 = train_X[0]

# Train data set Numbers
label = train_y[0]

# Set title
plt.xlabel("Le label  = "+ str(label))

# Show the image
imgplot =  plt.imshow(image_1)

# Show any other graph
# plt.show()


# Dimension of each items
print("X train = " + str(train_X.shape))
print("y train = " + str(train_y.shape))

print("X test = " + str(test_X.shape))
print("y test = " + str(test_y.shape))





```

```python
# Flatten the matrixe
x_train_flat = train_X.reshape(train_X.shape[0], -1)
x_test_flat =  test_X.reshape(test_X.shape[0], -1)


#Train the model
le = LinearRegression()
le.fit(x_train_flat, train_y)

# Predict
y_pred = le.predict(x_test_flat)

print(y_pred)
```

```python
#Changin model => Tree model
# modele = tree.DecisionTreeClassifier()
# modele  = RandomForestClassifier() #paramaters => n_estimators=100, max_depth=5, random_state=42
# modele = KNeighborsClassifier(n_neighbors=3)

# Current model 
modele = MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=10, random_state=42)

#Train the model
modele.fit(x_train_flat, train_y)


## Predict
y_pred = modele.predict(x_test_flat)

print("Done : " + str(y_pred));
```

```python
# Find the accuracy score
score = modele.score(x_test_flat, test_y) * 100
# With Accuracy
accuracy = accuracy_score(y_pred, test_y) * 100

# Display the decision tree !!! DO NOT WORK WITH RANDOM TREE , TOO MUCH 
# plt.figure(figsize=(200,100))
# tree.plot_tree(modele, filled=True, rounded=True, max_depth=5, class_names=[str(i) for i in range(10)])
# plt.title("Arbre de décision")
# plt.show()

print(f"Pourcentage de prédictions correctes : {accuracy:.2f}%")

```

```python
#Confusion matrix 

cm = confusion_matrix(test_y, y_pred)

disp = ConfusionMatrixDisplay(cm)
disp.plot()
plt.title("Matrice de confusion")
plt.show()
```

```python
# Charger une image depuis l'ordinateur
# Remplacez "chemin/vers/image.png" par le chemin réel de votre image
image_path = "4.png"
img = Image.open(image_path).convert('L')  # Convertir en niveaux de gris
img = img.resize((28, 28))  # Redimensionner l'image à 28x28 pixels
img_array = np.array(img).reshape(1, -1)  # Aplatir l'image pour qu'elle corresponde aux données d'entrée

# Prédire le label pour l'image importée
predicted_label = modele.predict(img_array)[0]


# Afficher l'image et la prédiction
plt.imshow(img, cmap='gray')
plt.title(f"Prédiction: {predicted_label}")
plt.axis('off')
plt.show()
```
